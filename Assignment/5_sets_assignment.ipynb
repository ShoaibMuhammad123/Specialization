
### Assignment 3: Set Operations

Create two sets: one with the first 5 positive integers and another with the first 5 even integers. Perform and print the results of union, intersection, difference, and symmetric difference operations on these sets.


```python
integer_set = {1,2,3,4,5}
even_set = {0,2,4,6,8}

## Union 
union_set = integer_set.union(even_set)

## Intersection 
intersection_set = integer_set.intersection(even_set)

## Difference 
difference_set = integer_set.difference(even_set)


## Symmetric Difference
symmetric_difference = integer_set.symmetric_difference(even_set)

print("Union result: ",union_set)
print("intersaction result: ",intersection_set)
print("Difference result: ",difference_set)
print("Symmetric difference result: ",symmetric_difference)
```


### Frozenset
* A frozenset is an immutable version of a set, meaning it cannot be changed after it is created. It is an unordered collection of unique elements, similar to a standard set, but with the key difference that you cannot add or remove items from it. Because they are immutable, frozensets are "hashable," which allows them to be used as keys in a dictionary or as elements within another set. 


### Assignment 9: Set and List Conversion

Create a set with the first 5 positive integers. Convert it to a list, append the number 6, and convert it back to a set. Print the resulting set.


```python
set1 = set(range(1,6))
print("set: ",set1)

# conversion into list 
lst = list(set1)
# adding element 
lst.append(6)

print("list: ",lst)

# again convert list to set 
set1 = set(lst)

print("final set: ",set1)
```

### Conclusion:
In Python, sets provide a powerful and efficient way to work with collections of unique items. They automatically remove duplicates, offer fast membership testing, and support useful mathematical operations such as union, intersection, and difference. Because sets are unordered and mutable, they are ideal for tasks where uniqueness and quick lookups are important. Overall, sets help simplify data handling and make operations on distinct elements both clean and efficient.
