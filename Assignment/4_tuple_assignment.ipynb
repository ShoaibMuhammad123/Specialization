Assignment 10: Tuple and List

* Create a tuple with the first 5 positive integers. Convert it to a list, append the number 6, and convert it back to a tuple. Print the resulting tuple.


```python
tup = (1,2,3,4,5)
print("Fist it is tuple",tup)
lst = list(tup)
print("Converted it to list: ",lst)
lst.append(6)
print("Appended a value 6 in it: ",lst)
tup = tuple(lst)
print("Convert back into tuple: ",tup)
```


### Assignment 15: Tuple Functions

Write functions that take a tuple and return the minimum, maximum, and sum of the elements. Print the results for a sample tuple.

```python
def func(tup):
    return max(tup),min(tup),sum(tup)

tup =tuple([1,2,3,4,5])
sample_tup = func(tup)
print(sample_tup)

print("Min",sample_tup[0])
print("Max",sample_tup[1])
print("Sum",sample_tup[2])
```
