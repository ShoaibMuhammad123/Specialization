Define a recursive function to calculate the nth Fibonacci number using memoization. Test the function with different inputs.


`With out Memoization Fabonachi Series`

```python
first_num = 1
second_num=1
num=6

result =0
seq = []

seq.append(first_num)
seq.append(second_num)

for i in range(1,num):
    
    result = first_num +second_num
    seq.append(result)
    
    first_num =second_num
    second_num =result
    
print(result)
print(seq)
```

```python
def fibonacci(n, memo={}):
    if n in memo:       # if we already know the answer
        return memo[n]  # return it quickly

    if n <= 1:
        return n

    # calculate and store the result
    memo[n] = fibonacci(n-1, memo) + fibonacci(n-2, memo)
    return memo[n]

fibonacci(12)
```

```python

memo = {}

def square(n):
    if n in memo:
        print('returning')
        return memo[n]
    
    print("Calculating...")  # Just to show effort
    memo[n] = n * n
    return memo[n]

print(square(5))  # Calculating... → 25
print(square(5))  # Returned instantly → 25
```

```python
# Normal method defining new list and storing item in that list then returing
def integer_key_value_pair(**kwargs):
    temp_dict = {}
    for key,value in kwargs.items():
        if type(value) ==int:
            temp_dict[key]=value
    return temp_dict



integer_key_value_pair(name='shoaib',age=22,marks=33,height=5.8,gender ='male')
```

Define a function that takes another function as a callback and a list of integers. The function should apply the callback to each integer in the list and return a new list with the results. Test with different callback functions.


```python
def callback_function(fun,lst):
    return list(map(fun,lst))

def square(a):
    return a**2

l1 = [1,2,3,4,5]
l2 = [11,12,13,14,15]

print(callback_function(square,l1))
print(callback_function(square,l2))
```

```python
## 2nd Method 
def apply_callback(callback,lst1):
    return [callback(x) for x in lst1]


print(apply_callback(lambda x: x**2,[3,4,5,5]))
print(apply_callback(lambda x: x+3,[3,4,5,5]))
```

Define a higher-order function that takes two functions, a filter function and a map function, along with a list of integers. The higher-order function should first filter the integers using the filter function and then apply the map function to the filtered integers. Test with different filter and map functions.


```python
def higher_order_function(filter_func,map_func,lst_integer):
    
    return [map_func(x) for x in lst_integer if filter_func(x)]
    
    
higher_order_function(lambda x: x%2==0,lambda x:x**2,[1,2,3,4,5,6,7,8,9])


```

```python
def compose(f,g):
    return lambda x: f(g(x))


f = lambda x: x+3
g = lambda x: x*2


h =compose(f,g)


print(h(22))
print(h(55))
print(h(23))
```

Define a function that takes a list of integers and returns their average. The function should handle any errors that occur (e.g., empty list) and return None in such cases. Test with different inputs.


```python
def average(lst):
    try:
        average = sum(lst)/len(lst)
        return average
    except ZeroDivisionError:
        raise None
average([3,4,5,6,7,8])
```

```python
def handling_error(lst):
    if len(lst)==0:
        return None
        
    
    sum=0
    for value in lst:
        sum+=value
    return sum

print(handling_error([]))
```

```python
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Test
fib_gen = fibonacci_generator()
for _ in range(10):
    print(next(fib_gen))
```

* Reverse list with out slicing 

```python
def reverse_list(lst):
    # Your code goes here
    start =0 
    end = len(lst) -1
    while start < end:
        lst[start],lst[end]=lst[end],lst[start]
        
        start +=1 
        end -=1
    return lst
    
print(reverse_list([1,2,3,4,5]))
```

* Rotate list k times

```python
def rotate_list(lst, k):
    # Your code goes here
    for i in range(k):

        last_element = lst.pop()  # Remove the last element
        lst.insert(0, last_element)  # Insert it at the front
    
    return lst

print(rotate_list([1,2,3,4,5],2))
```

* Words Frequency count

```python
def count_word_frequency(sentence):
    # Your code goes here
    frequency={}
    words = sentence.split()
    for word in words:
        if word not in frequency:
            frequency[word] =1
        else:
            frequency[word]+=1
    return frequency
print(count_word_frequency("hello world hello"))
```

* Merge three dictionaries into one dictionary and add the values of common keys

```python
def merge_dicts_with_overlapping_keys(dicts):
    # Your code goes here
    
    new_dict ={}
    for dic in dicts:
        for key , value in dic.items():
            if key in new_dict:
                new_dict[key] =new_dict[key]+value
            else:
                new_dict[key]=value
    return new_dict
    
    
    
    
print(merge_dicts_with_overlapping_keys([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}, {'c': 5, 'd': 6}]))

        
```
