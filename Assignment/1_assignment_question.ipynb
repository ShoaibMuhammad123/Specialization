**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = int(input('enter year :'))

if ((year % 4==0 and year % 100 !=0) or (year % 400 ==0)):
    print(year , 'is leap year')
else:
    print(year , 'is not a leap year')
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
# even on reversing the name if it give the same name that's known as palendrome

# palendrome if abcdcba  => civic -> civic 

name =input('Enter a name to check whether it is palendrome:')

if name == name[::-1]:
    print('Name is Palendrome')
else:
    print('Not a palendrome')
```
