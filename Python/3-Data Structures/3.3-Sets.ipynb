# **Sets**
* Sets are a built-in data type in Python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference, and symmetric difference.

#### 1- Creating a set

```python
## we use curly braces to create a set
first_set = {1,2,3,4}
print(first_set)
print(type(first_set))
```

```python
empty_set =set()
print(empty_set)
print(type(empty_set))
```

```python
my_set = set([1,2,3,4,5,6])    # this is creating set but it also doing conversion of list into set
print(my_set)
```

```python
## Sets remove the duplicate elements

remove_duplicate = set([2,2,2,2,5,4,6])
print(remove_duplicate)
```

#### 2- Basic set operations

```python
## 1- Adding and Removing Elements
my_set.add(7)
print(my_set)
my_set.add(7)    # it ignore it because in sets only unique value store not duplicate
```

```python
## Removing element from set
print(my_set)
my_set.remove(3)    # 3 is removed from the set
print(my_set)    
```

* Remove(item) , It can be used at that time when we want to remove an element if exist, but, it item not exist it raise error

```python
my_set.remove(8)    # as you can see here
```

* discard() , If we want that even if the element is not present in the set , we don't want to show an error then we use discard() function it remove the element if exist and do nothing if the element not exist

```python
my_set.discard(8)   # 8 is not present, that's why it ignore it not give an error
print(my_set)
```

```python
## pop method in set
removed_element = my_set.pop()   # it remove the first element from the set
print(removed_element)
print(my_set)
```

#### 5-Sets Methods

```python
### Counting unique words in text

text = "This is the class i am taking, it is not boring"

words = text.split()
print("Words are: ",words)
print(len(words))

## Convert list of words to set to get unique words
unique_words = set(words)

print("Unique words are: ",unique_words)
print(len(unique_words))
```
