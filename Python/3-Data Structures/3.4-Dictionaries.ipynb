### **1-Introduction to Dictionaries**
Dictionaries are unordered collection of Items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

### **2-Creating Dictionaries**

```python
## First we create an empty dictionary
empty_dict ={}     # for dictionary we use curly brackets/ braces
print(empty_dict)
print(type(empty_dict))
```

```python
## Another way to create an empty dictionary is by using dict() class
empty_dict = dict()
print(empty_dict)
print(type(empty_dict))
```

```python
## As we said dictionary is a collection of key-value pairs
student= {"Name":"Shoaib",
        "Age":22,
        "grade":24}

print(student)
print(type(student))
```

* When you could get and error

```python
## Single key always exist 
student= {"Name":"Shoaib","Age":22,"grade":24}

print(student)
```

```python
# If we use same key and assign different values then the 2nd value replace the first value the key remain same
student= {"Name":"Shoaib","Age":22,"Name":24}   # 24 will replace "Shoaib"

print(student)
```

### **3-Accessing Dictionary Elements**

* Two method to Access the Element of Dictionary 
    * Calling direct the key
    * using method provide by dictionary i.e get()

```python
## Accessing Dictionary Elements
student= {"Name":"Shoaib","Age":22,"grade":"A"}


# if i want to access the grade 
print("Grad is ", student['grade'])
# to access name 
print("Name is ",student['Name'])
```

* Another method to access element of Dictionary using the default get() method provide by dictionrary

```python
print(student.get("Name"))
print(student.get("grade"))
print(student.get("Age"))

# But if we try to access some element which is not available then
print(student.get("Gender"))  # it return None if such key not exist

## but if we want that even if key can't exist then give me another message
print(student.get("Gender","Not available key"))
```

### 4-**Modifying Dictionary Elements**

### **5-Common Dictionaries Methods**

```python
keys = student.keys()   ## getting all the keys
print("Keys are: ",keys)

values = student.values()    ## getting all the values
print("Values: ",values)

items = student.items()      ## getting all the key value pairs
print("Key value pairs: ",items)
```

##### **Shallow Copy**

```python
student['Age']=22
```

```python
## If we directly assign a dictionary to another variable then what happen

student_copy = student
print(student)
print(student_copy)

print("-------------After Changes---------------------")
## If we make changes  Age value in student dictionary 
student['Age']=55
print(student)
print(student_copy)
```

**So, changing of one can also occur in both of the dictionaries**

```python
## Shallow Copy 
student_copy = student.copy()
print(student)
print(student_copy)

# Make Changes in one 
student['Age']=55
print(student)
print(student_copy)
```

**Here you can see the changes in student dictionary cannot occur in student_copy dictionary**

### 6-**Iterating over Dictionaries**

* We can use loops to iterate through the dictionary, keys , values, or items

```python
## Iterating Over Keys
for keys in student.keys():
    print(keys)
```

```python
## Iterating Over Values
for value in student.values():
    print(value)
```

```python
## Iterating Over Key value Pairs
for key,value in student.items():    # .items() function return key-value pairs in the form of list of tuples i.e [(key,value),(key,value)]
    print(f"{key} : {value}")
```

### **7-Nested Dictionaries**

```python
students = {
    "student1":{'Name':'Altaf','Gender':'M','Age':33},
    "Student2":{"Name":"Munir","Gender":'M','Age':22}
}
print(students)
```

* Accessing Nested dictionaries elements

```python
## to get the name and age of student1
print(students["student1"]['Name'])
print(students["student1"]['Age'])

## To get the Gender of student2
print(students["Student2"]["Gender"])
```

```python
## three nested dictionaries
students_dict = {
    "student1":{"name":"shoaib","Subject_with_marks":{"cs":33,"DS":23}},
    "student2":{"name":"Amir","Subject_with_marks":{"cs":90,"DS":88}}
}
```

```python
print(students_dict["student1"]["Subject_with_marks"]["DS"])
print(students_dict["student2"]["Subject_with_marks"]["cs"])
```

**That's how we can Access values from a dictionary which is nested**

### **7-Iterating over Nested Dictionaries**

```python
for student_id,student_info in students.items():
    print(f"{student_id} : {student_info}")
```

```python
for student_id,student_info in students.items():
    print(f"{student_id} : {student_info}")
    for key,value in student_info.items():
        print(f"{key}: {value}")
```

### **8-Dictionary Comprehensions**

```python
squares = {i:i**2 for i in range(10)}
print(squares)
```

```python
## How can i find only the square of Even Numbers

## Conditional dictionary Comprehension

even = {i:i**2 for i in squares if i %2==0}
print(even)
```

# **Practical Examples**

* Use a dictionary to count the frequency of element in list

```python
lst = [1,1,2,3,4,3,2,1,3,2,5,6,4,3]

print(lst)
```

```python
frequency ={}
for number in lst:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

#### Merging Dictionaries

```python
a={'a':3,'b':33}
b={'c':34,'b':2}

merged_dict = {**a,**b}
print(merged_dict)
```
