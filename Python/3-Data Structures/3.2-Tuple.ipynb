#### **Introduction to Tuples**
* Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different.Tuple can also store different types of values

#### **Creating Tuples**

```python
## Creating Tuples
### we just need to use these to empty paranthesis

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
## Another way is to use tuple()  function or class
tup = tuple()
lis = list()

print(type(tup))
print(type(lis))
```

```python
## Adding elements in these tuples
numbers = tuple([3,4,5,6])              # list can be converted to tuple and vice versa
print(numbers)
```

```python
## tuple can also store different datatype of values

mixed_tuple = (1,"Hello World",3.24,True)

print(mixed_tuple)
```

#### **Accessing Element from tuple**

* Indexing and Slicing in tuple is almost similar like list 

```python
print(numbers)
print(numbers[0])
print(numbers[3])
print(numbers[-1])
```

```python
## Slicing 
print(numbers[1:4])
print(numbers[-2:])
print(numbers[::])
```

#### **Tuples Operations**

```python
## Tuple support various Operations  
print(numbers," ",mixed_tuple)


## Concatination of two tuples --> make it a single tuple
concatinated_tuple = numbers + mixed_tuple
print(concatinated_tuple)
```

```python
mixed_tuple * 3
```

```python
print(numbers)
print(numbers * 4)
```

#### **Immutable Nature of Tuple**

```python
## Tuples are immutable(means value can't be changed)

# list is mutable see this 
lis = [2,3,4]
print('Original list items: ',lis)
lis[1]=1000
print('Changed list items: ',lis)

# Tuples are immutable 
tup = (2,3,4)
print('Original Tuple items: ',tup)
tup[1]=1000                             ## This show an error because it is immutable, it can't be changed once value assigned
print('Changed list items: ',tup)  
```

#### **If we need to changed the tuple item then first we convert it to list and then changed the value and again convert it to tuples**

### **Tuple Methods**

```python
## Tuple Methods    ---> Two common methods
numbers =(1,1,1,3,4,3,5,6,7)
# 1-count
print(numbers.count(1))

# 2-index of return index of an item in tuple
print(numbers.index(4))
```

#### **Packing and Unpacking Tuples**

**Packing**
* Packing refer to when we assign multiple values to a single variables, it is known as packing, these values can be stored in a form of tuple

* variable_name = value1,value2,value3 ...valueN
* i.e
* num = 1,"Shoaib",3.55,True


```python
packed_tuple = 1,"Hello",3.22,True   # packing 
print(packed_tuple)
print(type(packed_tuple))
```

**Unpacking**
* Unpacking refer to we assign the packed tuple /variable to that number of variable which has values in tuple


```python
var1,var2,var3,var4 = packed_tuple

print(var1)
print(var2)
print(var3)
print(var4)
```

```python
## Another Technique of Unpacking
first,*middle,last=packed_tuple
print(first)
print(middle)
print(last)
```

#### **Nested Tuples**

```python
## Nested List
lst = [[1,2,3,4],[6,7,8,9],[1,'Hello',3.12,"c"]]
print(lst)

# to pick the value 4 in nested list
print(lst[0][3])

## Slicing can also be applied in Nested List
print(lst[1][2:])
```

```python
## Nested Tuples
lst = ((1,2,3,4),(6,7,8,9),(1,'Hello',3.12,"c"))
print(lst)

# to pick the value 4 in nested list
print(lst[0][3])

## Slicing can also be applied in Nested List
print(lst[1][2:])
```

```python
## Iterating through the Nested tuple using for loop

for i in lst:
    for j in i:
        print(j,end=" ")
    print()
```
