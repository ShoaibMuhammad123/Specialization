# **DataTypes**
* Each programming has data types, now we see with how many data types python is working with.

```python
# Integer
print(9)
```

```python
# how much large Integer python can handle  = 1 x 10^(309)
print('this is defined: ', 1e308)
print('this is undefined: ',1e309)
```

```python
# Decimal/ Float
print(8.33)


# max number 
print('Defined: ', 1.7e308)
print('Undefined: ',1.7e309)
```

```python
# Boolen 
print(True)
print(False)
```

```python
# Text/ String
print('Hello')
```

```python
# Complex Number
print(5+3j)
```

```python
# list
print([3,4,5,5])
```

```python
# Tuple
print((3,4,5))
```

```python
# Sets
print({3,4,5})
```

```python
# Dictionaries
print({'name':'shoaib','age':34})
```
