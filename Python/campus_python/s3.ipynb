### **1-Creating Strings**

```python
### first method is using single inverted quotes
s = 'Shoaib'
print(s)
print(type(s))

### Use can also use double inverted comma or quotes
s = "shoaib"
print(s)
print(type(s))

### you can also use triple inverted commas or quotes
s = '''Shoaib Muhammad'''
print(s)
print(type(s))

### Another technique is to use str() function
s = str("I am shoaib Muhammad")
print(s)
print(type(s))
```

* Their is a reason of three different techniques because if you have this sentence given below:
* **That's is a good approach** = Here you can't use a single inverted comma 

```python
string = "That's how u can see "   # here single inverted comma cannot use because it contain in the sentence
print(string)

string = """ Ahmad says that's how "Might is right" """  # here u can't use single or double inverted comma

print(string)
```

### **2-Accessing substring from string**

##### Accessing string can be done through **Indexing**
**Indexing has two types**
1. Positive Indexing
2. Negative Indexing

Through indexing we can get only a **single character**

```python
# Positive Indexing --> Python by default assign index to each character you u create string. 
#                       In positive indexing it start from left to right from 0 
s = "Hello World"

# you can access each character
print(s[0])   # to get H
print(s[9])
# print(s[11])   # it give you can error the index value is not found
```

```python
# Negative Indexing --> from right to left start from -1

print(s[-1])   # to print the last element
print(s[-2])   # to print the 2nd last element
```

#### To get multiple character we use **Slicing.**

* In slicing you can extract multiple character at a single time

```python
s = "Hello World"

## to get "Hello" onlly
print(s[0:5])
```

```python
print(s[:3])
print(s[-3:])
print(s[2:3])
```

```python
# to jump two character 
print(s[1:8:2])
print(s[-6:-1:2])
```

```python
## to reverse the given string
print(s[::-1])
```

### Function only For **Strings**

* **Capitalize/Title/Upper/Lower/Swapcase**

```python
s = "shoaib Muhammad"

print(s.capitalize())   # it convert your sentence first character into capital

print(s.title())     # capitalize each first letter  of each word

print(s.upper())     # cappitalize all the characters

print(s.lower())     # convert the character in to lower case

print(s.swapcase())   # It convert the capital letter to lower and vice versa
```
