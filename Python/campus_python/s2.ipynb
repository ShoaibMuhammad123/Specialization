##### Guessing Game

```python
# In this game user guess a number then computer give hint like guess below number
# or guess higher then user again guess then computer give some hint again untill they become same

# there is a problem it only execute once 

import random

jackpot = random.randint(1,100)
guess = int(input("Guess a number"))

if guess < jackpot:
    print('Incorrect! guess higher')
else:
    print('incorrect! guess hower')
```

```python
# to solve the above problem of the game we need a loop
jackpot = random.randint(1,100)
guess = int(input("Guess a number"))
counter = 1
while guess !=jackpot:
    

    if guess < jackpot:
        print('Incorrect! guess higher')
    else:
        print('incorrect! guess hower')
    guess = int(input("Guess a number"))
    counter +=1    
else:
    print("Correct Guess")
```
