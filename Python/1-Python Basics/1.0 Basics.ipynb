# Syntax
Syntax is the set of rules that define how symbols, words, or statements must be arranged so that they are valid in a language (like Python, English, or math).

# Semantics
Semantics is about meaning — what the code does when it runs, or what the statement means.

Even if code is syntactically correct, it might be semantically wrong — meaning it doesn’t do what you intended.

**x = 10 / 0**

**age = "twenty\"
print(age + 1)**

# Indentation
Indentation in Python refers to the spaces or tabs placed at the beginning of a line of code to define the block structure of the program.


```python
Age=10
if Age >4:
    print('Age is greater')
```

# Case Sensitivity
Python is a case sensitive programming language, it means does not know that 'Capital and Small' letters in Alphabets are same, it consider them different

**A** is different than **a**

**Pakistan** is different than **pakistan**

```python
name="Shoaib"
Name = "Muhammad"

print(name)
print(Name)
```

# Line Continuation
If our expression becomes very large and we want to continue that same expression in next line we use line Continuation. For line continuation we use back slash \

```python
# here it continue the line
total_marks = 4+2+5+7+44+63+\
    2+4+5
    
print(total_marks)
```

# Key Takeaways
* Syntax in Python refers to the correct arrangement of words and symbols, while semantics refers to the meaning and interpretation of code.

* Python is case sensitive, and variable names with different cases are treated as different identifiers.

* Indentation is crucial in Python to define code blocks, replacing the need for braces used in other languages.

* Python supports single-line and multi-line comments, line continuation with backslash, and allows multiple statements on a single line using semicolons.
