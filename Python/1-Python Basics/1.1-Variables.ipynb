* In Python Variables are created when you assign a value to them, and they donot need any explicit declaration to reserve memory space. 

* The declaration happen atomatically when you assign a value to a variable.

### Understanding Variables types

```python
# Python is dynamically typed : It mean that the type of variable is \
                                # determined at run time
age = 23  # integer
height = 5.9  # float
name = "Shoaib"  # String
is_student = True # bool


print(type(is_student))    # to check the type of variable 
```

### Type Checking and Conversion

```python
type(height)
```

```python
# to convert one data type into another data type

age = 55
print(age)
print(type(age))

# Type conversion

# we convert this age into string 
age_str = str(age)
print(age_str)
print(type(age_str))

```

```python
# can we are able to convert string to integer
age = '33'    # here inside quotation their is a valid numeric number so we can convert 
print(age)
print(type(age))


age = int(age)
print(age)
print(type(age))
```

```python
# Can we convert all strings into Number 
#    The answer is NO

name  = "Pakistan"    # we can't convert because it has no valid numeric number inside ''
# int(name)
```

```python
# Converting float into integer 

height = 5.9
print(height)
print(type(height))

print('-'*20)
# Casting 
height = int(height)
print(height)
print(type(height))
```

```python
# Type Casting from int to float 
marks = 40
print(marks)
print(type(marks))

print('-'*20)

marks = float(marks)
print(marks)
print(type(marks))
```

### Dynamic Typing
* Python alows the type of variable to change as the program executes


```python
## Dynamic typing
## Python alows the type of variable to change as the program executes
var = 10 # int
print(var,type(var))

var = 'hello'
print(var,type(var))

var = 3.4
print(var,type(var))
```

```python
## input 
# we want from user to enter input
user = input('Enter Numer please')
print(user)
print(type(user))    # even if we enter number the input function convert its type into string 

```

```python
# if we want to get the number in integer form
user = int(input('Enter input please'))
print(user,type(user))
```
