# **DataTypes**

## 1.Definition:
* Data Types are the classification of data which tell the compiler or interpreter how the programmer intends to use the data
* They determine the types of operations that can be performed on the data, the values that the data can take, and the amount memory needed to store the data.

* **What types of data programmer wants to store.**
* **On basis of data types we find what types of operations we can apply on it.**
* **Also, we can find how much memory it takes**

### 2. Importance of Data Types in Programming

#### Explanation
* Data Types ensure that data is stored in an efficient way.
* They help in performing correct operations on data.
* Proper use of data types can prevent errors and bugs in the program.

Given:

a = [1, 2, 3]

b = a

b += [4, 5]

print(a)

* A. [1, 2, 3]
* B. [1, 2, 3, 4, 5]
* C. Raises Error
* D. [4, 5]

Answer: B

Explanation: += modifies the list in place; both a and b refer to the same object.

```python
a = (1, 2)
b = a
b += (3, 4)
print(a)
```

Now compare:

a = (1, 2)

b = a

b += (3, 4)

print(a)


* A. (1, 2)
* B. (1, 2, 3, 4)
* C. Raises Error
* D. Implementation-dependent

Answer: A

Explanation: Tuples are immutable; b += (3, 4) creates a new tuple, leaving a unchanged.
