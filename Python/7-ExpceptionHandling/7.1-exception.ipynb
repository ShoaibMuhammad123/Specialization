## `Understanding Exception`
* Exception handling in Python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This lesson will cover the basics of exceptions, including how to use try, except,else and finally block.

### `What are Exception?`

Exceptions are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. Common exceptioons include:

* **ZeroDivisionError**: Dividing by zero
* **FileNotFoundError**: File not found
* **ValueError**: Invalid value.


`Error`
* An error generally refers to a problem in the code itself — often something that cannot be handled at runtime.
* Examples:
    * **Syntax errors**
    * **Indentation errors**
    * These stop the program before it even runs

`Exception`
* An exception is a runtime issue — something that goes wrong while the program is running — and can be handled with try/except.
* Examples:
    * **ZeroDivisionError**
    * **ValueError**
    * **TypeError**
    * **FileNotFoundError**

```python
a = b   # this NameError is one type of exception
```

```python
## How do i handle this 
try:   # inside this block i will write a block of code their exception may raise
    a=b 
except:
    print("The variable has not been assigned")
```

```python
## to display the message of error 
try:
    a=b 
except NameError as ex:
    print(ex)    ## This is the particular error message
```

```python
try:
    a=b
except Exception as ex:
    print(ex)
```

```python
### similarly other example
## ZeroDivisionError is also a kind of exception and belong to Exception class
try:
    result = 1/0
except ZeroDivisionError as ex:
    print(ex)
    print('please enter the denominator greater than zero')
```

* **Some time their may be different type of exception but, we can write code to handle all of them, To sollve this problem what we can do is we use the Parent Exception class**

```python
try:
    result = 1/3
    a = b 
except ZeroDivisionError as ex:
    print('please enter number greater than zero')
    
except Exception as e:    # this is the base class of Exception (Every exception is derived from this class)
    print(e)
    print('Parent class has handled the error')
    
```

* Remember the base exception caught over here 

```python
try: 
    num = int(input('enter a number'))
    result  = 2/num
except ValueError as ex:
    print(ex)
except ZeroDivisionError as ex:
    print(ex)
    print('Inside zero division exception handling')
    
except Exception as ex:  # Base model
    print(ex)
```

`try,except,else`

`else`
* else block get execute successfully only if the `try` block has executed successfully

```python
try: 
    num = int(input('enter a number'))
    result  = 2/num
except ValueError as ex:
    print(ex)
except ZeroDivisionError as ex:
    print(ex)
    print('Inside zero division exception handling')
except Exception as ex:  # Base model
    print(ex)
else:
    print("The try block has executed successfully")
    print('And the result is:',result)
```

`try, except, else, finally`

`finally`
* Whether the error is comming or not it executed all the time
* In case of 'else' block only execute if the try successfully execuate with out occuring of error

```python
try: 
    num = int(input('enter a number'))
    result  = 2/num
except ValueError as ex:
    print(ex)
except ZeroDivisionError as ex:
    print(ex)
    print('Inside zero division exception handling')
except Exception as ex:  # Base model
    print(ex)
else:
    print("The try block has executed successfully")
    print('And the result is:',result)
    
finally:
    print("The execution has completed ")
```

* Practical Example

```python
## File handling and Exception handling

try:
    file = open('example.txt','r') 
    content = file.read()
    print(content)
except FileNotFoundError as ex:
    print(ex)
    
else:
    file.closed()
    
finally:
    print('Excecution complete.')
    
```
