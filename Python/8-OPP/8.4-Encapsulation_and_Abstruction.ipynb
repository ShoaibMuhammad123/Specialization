# **Encapsulation And Abstruction**

Encapsulation and abstraction are twon fundamental principle of Object-Oriented Programming(OOP) that help in designing robust, maintainable, and reusable code. Encapsulation involves bundling data and methods that operate on the data with a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary features.

# **Encapsulation**
Encapsulation is the concept of wrapping data(variables) and methods (functions) together as a single unit. It restricts direct access to some of the object's components, which is a mean of preventing accidental interference and misuse of the data.

* `Public variable`

```python
### Encapsulation with Getter and Setter Methods

### Public, Protected, and Private variables

class Person:
    def __init__(self,name,age):
        self.name = name    ## public variable (mean I can call or use this variable out of this class )
        self.age = age      ## public variable


person = Person('shoaib',33)
person.name   ## Here outside the class the name and age are accessible
```

```python
## to check all the public variable for this 'person' object

dir(person)
```

* If I have to define the variable and method in such a way that outside the class it can't be accessible then we do this see below

`Private Variables`
* Private variable can be formed by placing double underscore sign with the starting of a variable. It cannot be access outside the class and also not be access in derived class or child class

```python
class Student:
    def __init__(self,name,age):
        self.__name  = name   ## self.__name is private
        self.__age = age      ## self.__age is private
        
student = Student('Shoaib',22)

## now we can't access this name because it is private    

student.__name   # It give an error because it is private
```

```python
## If we try to check the attribute and method now 
dir(student)   ## here you can see that their is no name attribute like before
```

`Protected Variable`
* Protected Variable cannot be access outside the class but these variable can be access in a `derived class /child class` 

```python
class Mobile:
    def __init__(self,brand,model):
        self._brand = brand    ## self._brand is private variaable
        self._model = model    ## self._model is private variable
        
class Iphone(Mobile):
    def __init__(self, brand, model,camera):
        super().__init__(brand, model)
        self._camera = camera
        
iph = Iphone('apple',2022,500)

iph._model
    
```

`Encapsulation with Getter and Setter`
* 

```python
class Animal:
    def __init__(self,name,age):
        self.__name = name    ## private access modifier or variable
        self.__age = age      ## private variable
        
    ## Getter method for name 
    def get_name(self):
        return self.__name
    
    ## Setter method (it help us to change the value)
    
    def set_name(self,new_name):
        self.set_name = new_name
        
    ## Getter method for name 
    def get_age(self):
        return self.__age

    
    ## Setter method (it help us to change the value)
    
    def set_age(self,new_age):
        if new_age < 0:
            print('Age can not be negative')
        else:
            self.__age = new_age
           
dog = Animal('tomy',22)

 ## accessing the variable through a proper way
print(dog.get_name())
 
## Accessing and modifying using getter and setter

print(dog.get_age())

dog.set_age(64)
print(dog.get_age())

dog.set_age(5)
print(dog.get_age())
```
