## **Inheritance**


* Inheritance is a fundamental concept in Object-Oriented Programming (OOP) that allows a class to inherit attributes and methods from another class. This lesson covers `single inheritance` and `multiple inheritance`, demonstrating how to create and use them in Python

`Single Inheritance`

* In single inheritance a child class can only inherit from a single parent class. 
* For single parent class inheritance we use `super()` keyword. 
* For Mulitple class inheritance we use `Class Name` to inherit 

```python
## Single Inheritance

class Person:  
    def __init__(self,name,age,gender):
        self.name = name
        self.age = age
        self.gender = gender
        
    def eat(self):
        print('Person can eat')
        
class Student(Person):
    def __init__(self,name,age,gender,marks):
        super().__init__(name,age,gender)
        
        self.marks = marks
        
    def sing(self):
        print('Student can sing.')
        

std = Student('Shoaib',22,'M',33)
std.eat()     
std.gender
              
```

* Multiple Inheritance mean that a chield class can inherit the two parent classes.

```python
## Multiple Inheritance 

class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    def walk(self):
        print('A person can walk')
        
class Profession:
    def __init__(self,prof_name):
        self.prof_name = prof_name
        


class Teacher(Person,Profession):
    def __init__(self, name, age,prof_name):
        Person.__init__(self,name,age)
        Profession.__init__(self,prof_name)
        
    def taught_student(self):
        print('Teacher can taught student')
        
        
teacher1 = Teacher('Shoaib',22,'Lecturer')

teacher1.taught_student()
teacher1.name

        
        
        
```

**Summary**
* Single inheritance allows a child class to inherit from one parent class.
* Multiple inheritance allows a child class to inherit from multiple parent classes.
* The super() function is useful for calling the parent class constructor in single inheritance.
* In multiple inheritance, parent class constructors can be called explicitly using the class name.

**Key Takeaways**
* Inheritance in Python allows a class to inherit attributes and methods from another class, facilitating code reuse.
* Single inheritance involves one child class inheriting from one parent class, demonstrated with a Car and Tesla example.
* The super() function is used in child classes to call the parent class's constructor and inherit its attributes.
* Multiple inheritance allows a class to inherit from more than one base class, as shown with Animal, Pet, and Dog classes.
