## **Class and Objects**

* `OPP` stands for Object Oriented Programming is a programming paradiagm that uses `Objects` to design applications and computer programs. OPP allows for modeling real-world scenarios using classes and objects. This lesson covers the basics of creadting classes and objects, including `instance variable and methods`

 programming.

`Instance Variable and Methods`

```python
## Instance variable and methods

class Car:
    ## Constructor  ---> Inside it we can define all our instance variable/attribute/property
    
    def __init__(self):   ## self refer to the current object(you can use you own name also, but self is using conventionally)
        ## Instance variable are the propertise about some particular object in this case(car)
        self.door =2
        self.wheel = 4
        self.color ='red'
        
        ## Now let's talk about instance method 
        


## Create the object
bmw = Car()

print(bmw)
print(bmw.wheel)
print(bmw.color)
print(bmw.door)
```

```python
class Dog:
    def __init__(self,name,color,age):
        self.name = name
        self.color = color
        self.age =age
        
    def bark(self):
        print(f'{self.name} is barking, \nand its age is {self.age}')
    
dog1 = Dog('tomy','white',10)
print(dog1.name)
print(dog1.color)
print(dog1.age)

## Accessing method 
dog1.bark()

dog2 = Dog('jerry','red',24)

dog2.bark()
    
```

### `Modeling a Bank Account`

```python
class Bank_Account:
    def __init__(self,balance =0,money=0):
        self.balance = balance
        self.money = money
        
    def display_balance(self):
        print(f'Total balance is: {self.balance}')
        
    def Add_Amount(self,money2):
        self.balance += money2
        print(f'Amount {money2} is deposited, and current balance is {self.balance}')
        
    def With_draw(self,amount):
        if self.money < self.balance:
            self.balance -= self.money
            print(f'Total Remaining Balance:\n{self.balance}')
            print(f'This {amount} has been withdrawn successfully')
            print('New balance is {}'.format(self.balance))
        else:
            print('Insufficient Funds')
acc1 = Bank_Account(200,200)

# acc1.display_balance()
acc1.Add_Amount(56)
acc1.display_balance()
acc1.Add_Amount(3000)
acc1.With_draw(33)

            
        
```
