`Joining Paths`

```python
### Joining path very important concept

dir_name = "folder"
file_name='file.txt'
full_path = os.path.join(dir_name,file_name)

print(full_path)
```

`If I want to get my full path`

```python
### Joining path very important concept

dir_name = "folder"
file_name='file.txt'
full_path = os.path.join(os.getcwd(),dir_name,file_name)

print("This is the full path",full_path)
```

`To conform that is the path exist or not`

```python
path = 'example.txt'
if os.path.exists(path):
    print(f'the path "{path}" is exist')
else:
    print(f"the path {path} doesn't exist")
```

```python
path = 'simple.txt'
if os.path.exists(path):
    print(f'the path "{path}" is exist')
else:
    print(f"the path '{path}' doesn't exist")
```

#### Checking `Whether the path is a file or directory`

```python
path = "example.txt"
if os.path.isfile(path):
    print(f"the '{path}' is a file")
    
elif os.path.isdir(path):
    print(f"the '{path}' is a directory")
else:
    print(f" the {path} is neither a directory nor a file")
```

```python
### Getting the absolute path of a folder

relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)

print('relative path of example.txt',relative_path)   # it gives us a relative path
print('Absolute path of example.txt',absolute_path)   # it gives us a complete path
```
