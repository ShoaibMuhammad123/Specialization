# **File Operation-Read And Write Files**

`File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.`

* `Reading the entire file content`

```python
## Read a whole file
with open('example.txt','r') as file:
    content = file.read()
    print(content)   # with this i am reading the entire content
        
```

* `To read the file line by line.`

```python
with open('example.txt','r') as file:
    for line in file:        ## file is working as a iterator
        print(line)        ## It print an extra space in b/w the lines
```

```python
## To remove the extra line b/w the lines we use ".strip()"
with open('example.txt','r') as file:
    for line in file:        
        print(line.strip())        ## .strip remove the new line character
```

* `To open a file for writing something init.`

- 'File Overwriting'

```python
## writing in a file(Overwriting) 

with open('example.txt','w') as file:
    file.write('Hello World!\n')
    file.write('this is a new line.')

## This code will remove the entire content and then write these text over that file
```

* `Writing in a file without overwriting`

```python
with open('example.txt','a') as file:
    file.write('\nAppend operation taking place.')
```

* `Their may be the scenario where I have to write multiple lines.`

```python
### Writing list of lines to a file

lines =['\nFirst line \n','Second line \n','Third Line \n']

with open('example.txt','a') as file:
    ## inorder to write all the lines in side this list  to a file 
    ## we use a function called file.writelines()
    file.writelines(lines)
```

* The above are wrt text file.

* `Now we discuss with respect to binary file`
    - Binary file has `.bin` extension
    - In binary files we usually update bytes

`Writing into a binary file`

```python
### Binary files

# Writing to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

`Reading a binary file`

```python
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

#### `Now discuss some practical Examples`

* If i want to read the content from a `source text file` to a `destination text file`

```python
## copying a text file from source.txt to destination.txt file

with open('example.txt','r') as source_file:
     content = source_file.read()
     
with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```

`Assignment:`Read a text file then count the number of lines, words and characters

```python
with open('example.txt','r') as file:
    lines =0
    words =0
    characters =0
    for line in file:
        lines +=1
        words +=len(line.split())
        characters +=len(line)
        
print('Line',lines)
print('Words',words)
print('characters',characters)
```

`The w+ mode in python is used to open a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its content is truncated (i.e., the file is overwritten)`

```python
### Writing and then reading a file

with open('example.txt','w+') as file:
    file.write("Hello World\n")
    file.write("This is a new line \n")
    
    # Moving the cursor to the begining  we used file.seek() --> it move the cursor to the begining
    file.seek(0)
    
    ## Read the content of the file
    content = file.read()
    print(content)
```
