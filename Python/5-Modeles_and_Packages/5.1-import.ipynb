# **Import Modelues in Python: Modules and Packages**
* Modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

#### Syntax
`import package_name` **For importing all the function from the packages**

`from package_name import function_name`  **To import a specific function from a package**

```python
import math
math.sqrt(25)
```

```python
# to get only a single function from math package
from math import sqrt

sqrt(36)
```

```python
math.ceil(3.6)   # it increase the number to integer 3.3--> 4
```

```python
math.ceil(3.3)
```

```python
# to decrease to a number to nearest integer
print(math.floor(3.3))   
print(math.floor(3.7))
```

```python
print(math.pi)
```

```python
print(math.inf)
```

```python
print(math.cos(360))
```

# `How do I create MY own custom Package`

**To create a custom package we have to create a `package`folder and inside that a `__init__.py()`\
 then another python file inside this file all the function we want to write.**

```python
## How to perform addtion by importing the custom created package 
## and function

from package.maths import addition   # importing custom package


result = addition(2,3)    # custom addition function
print(result)
```

```python
from package import maths

maths.addition(4,5)
```
