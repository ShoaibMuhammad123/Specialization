* `random()`: Returns a random floating-point number between 0.0 (inclusive) and 1.0 (exclusive). 
* `uniform(a, b)`: Returns a random floating-point number between a and b (inclusive). 
* `randint(a, b)`: Returns a random integer between a and b (both inclusive). 
* `randrange(start, stop, step)`: Returns a randomly selected integer from the range specified by start, stop, and step. 

```python
## Random modulle
import random

print(random.randint(10,20))   # return a value randomly between the range a,b in this case 10,20 both included
print(random.random())          # return a floating point random number between 0.0(inclusive) and 1.0(exclusive)
print(random.uniform(3,4))
print(random.choice(['a','b','c']))    # select a single element randomly
print(random.sample(range(100),k=20))

```

```python
random.sample(range(10,20),k=9)
```

```python
random.randrange(2,10,2)
```

`Data Serialization`
* Data serialization is the process of converting a `data object` (such as a Python dictionary, a Java object, or an in-memory data structure) into a format that can be stored, transmitted, or reconstructed later.


A `data object` is:

* A value with a specific type (e.g., number, string, list, dictionary, class instance)

* Stored in memory

* Manipulated by code

```python
import json
data ={'name':'shoaib','age':23}

json_str = json.dumps(data)
```

```python
print(json_str)
print(type(json_str))   ## This is the serialized form of json
```

```python
## To convert this back again into dictionary
data_json = json.loads(json_str)
```

```python
print(data_json)
print(type(data_json))
```

```python
## CSV file

import csv

with open('example.csv',mode='w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['krish',23])
    
with open('example.csv',mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
