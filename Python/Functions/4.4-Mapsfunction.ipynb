# **The Map() Function in Python**

* The map() function applies a given function to all items in an input list (or any other iterable) and return a map object(an iterator). This is particularly useful for transforming data in a list comprehensively.

```python
## If i have to square a number then function would be 
def square(n):
    return n**2

square(4)
```

```python
## If i have a list of Numbers 
numbers = [1,2,3,4,5,6,7,8]

## To perform the square operation on this list of number  , we have to perform the following operation

# we have to create empty list
# then define a function 
# inside a function we have to run a loop
# then we have to append the square of a number in a list

# now we apply

square_list=[]
def square(num_list):
    for num in num_list:
        square_list.append(num**2)
        
square(numbers)
print(square_list)
```

### **Syntax of map() functinon**
`map(function,iterable)`

`Instead of doing these we directly use map() function that make it easy`
* this is how **map()**  make it easy

```python
numbers = [1,2,3,4,5,6,7,8]

list(map(square,numbers))     # this is one way
```

`map() function with lambda function `

```python
list(map(lambda num:num**2,numbers))   # this is how map() and lambda function make it easy
```

### **Can we map multiple iterables**

```python
numbers1 = [1,2,3]
numbers2 = [4,5,6]

addition = list(map(lambda x,y:x+y,numbers1,numbers2))
print(addition)
```

```python
### How to add three list with corresponding element

l1 =[3,4,5]
l2 =[5,4,3]
l3 =[2,2,2]

add_result = tuple(map(lambda x,y,z:z+y+z,l1,l2,l3))
print(add_result)
```

`3-We can also use map() function to convert list of **Strings** to integers`

```python
str_numbers = ['1','2','3','4','5']
int_numbers = list(map(int,str_numbers))    # We are doing type casting over here

print(int_numbers)
```

`4-To make upper all the letters from a string list`

```python
fruits = ['apple','banana','mango','orange']

upper_fruits = list(map(str.upper,fruits))
print(upper_fruits)
```

`5-To apply on a list of dictionary`

```python
def get_name(person):
    return person['name']

get_name({'name':'shoaib','age':33})
```

```python
persons =[
    {'name':'shoaib','age':22},
    {'name':'Ali','age':34},
    {'name':'Munir','age':54}
]

list(map(get_name,persons))
```

`Conclusion`
* The map() function is a powerful tool for applying transformations to iterable data structures. It can be used with regular functions, lambda functions, and even multiple iterables, providing a versatile approach to data processing in Python. By understanding and utilizing map(), you can write more efficient and readable code.

# **Top 10 Interview Questions on Lambda Functions in Python**

`1. What is a lambda function in Python? How is it different from a normal function?`

* Explain that lambda is an anonymous, single-expression function defined using lambda.

* Key differences: no name, limited to one expression, often used for short inline operations.

`2. Write a lambda function to add two numbers. How would you call it?`



```python
add = lambda a, b: a + b
print(add(2, 3))
```

`3. In what scenarios would you prefer lambda functions over regular functions`
* Used with **map**, **filter**, **reduce**, **sorting keys, callbacks, inline transformations**, etc.

`4. What are the limitations of lambda functions in Python?`

* Only one expression allowed

* No statements, loops, annotations, or complex logic

* Harder to debug

* Not easily reusable compared to named functions

```python
[lambda x, i=i: x + i for i in range(3)]
```

```python
[(lambda x: x*x)(i) for i in range(11) if i % 2 == 0]
```
