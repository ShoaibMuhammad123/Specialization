### `1-Example1: Temperature Conversion`

```python
# write two functions that convert celcius to fahrenheit anv vice versa
def celcius_to_fahrenhiet(celcius_temp):
    fahrenheit = (celcius_temp * 9/5) + 32
    return fahrenheit

celcius_to_fahrenhiet(25)
```

```python
def fahrenheit_to_celcius(fah):
    celcius = (fah - 32) * 5/9
    return celcius
    
print(fahrenheit_to_celcius(77))
```

```python
def temperature_conversion(temp,unit):
    
    """
    The fumctopm convert celcius to 
    fahrenheit and fahrenheit to celcius
    """
    
    if unit =="Celcius":
        print("The temperature is in Celcius, We convert it in Fahrenheit")
        fah = temp * 9/5 +32
        return fah
    else:
        print("The temperature is in Fahrenheit, We convert it in Celcius")
        return temp-32 * 5/9
    
temperature_conversion(25,'Fahrenheit')
    
```

### `2-Example2 : Word Strength Checker`

```python
def is_password_strong(password):
    """This can check that whether the password is strong or weak """
    
    if len(password) < 8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char in '@#$%^&*!' for char in password):
        return False
    return True
```

```python
print(is_password_strong("ShoaibMuhammad"))
print(is_password_strong("Shoaib"))
print(is_password_strong("StrongPassword45"))
print(is_password_strong("Stronglife12@1"))
```

#### `3-Example3-Calculate the total cost of Items in a Shopping Cart`

```python
def total_cost_of_items(cart):
    """This function calculate the total cost of items in a shopping cart"""
    total_cost=0
    for item in cart:
        total_cost += item['price'] * item['quantity']
    return total_cost
        

# Example of cart data
cart =[
    {"name":"apple","price":0.3,"quantity":4},
    {"name":"Banana","price":0.4,"quantity":2},
    {"name":"Mango","price":0.2,"quantity":6},
]


# calling the total_cost function
total_cost_of_items(cart)
```

#### `4-Example4 Check if a String is Palindrome`
* Palendrome mean even if we reverse that the string or Name remain the same

```python
def palendrome(input_str):
    """This function can check wheter the string is palendrome or not"""
    input_str = input_str.lower().replace(" ","")
    if input_str==input_str[::-1]:
        return "The Given String is Palendrome"
    
## Calling the function
palendrome("Civic")
```

#### `Example5-Calculate the factorial of a number using recursion`

```python
def factorial(n):
    """This function return the factorial of a given number using recursion"""
    if n==0:
        return 1
    else:
        return n* (factorial(n-1))

print(factorial(5))
```

```python
def read_file(file_path):
    
    frequency ={}

    with open(f'{file_path}','r') as file:
        # fist_line = file.readline()
        all_text = file.readlines()

        for text in all_text:
            split_text = text.split()
            for word in split_text:
                if word in frequency.keys():
                    frequency[word]+=1
                else:
                    frequency[word]=1
        return frequency
        
read_file('sample.txt')
    
    
```
