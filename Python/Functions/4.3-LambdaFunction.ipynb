# **Lambda Functions in Python**
* Lambda functions are small anonymous functions defined using the lambda keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions

1. **Anonymous function** mean function with out a name.
2. To define it we have to use **lambda** keyword
3. It should have one expression



### **Syntax**
`lambda arguments: expression`

```python
## If we define a basic function
def addtion(a,b):
    return a+b
addtion(2,3)
```

* If i create a `Lambda function` for this basic addtion

```python
addition =lambda a,b: a+b
print(type(addition))

## In order to class this lambda function 
print(addition(4,2))
```

* To Create a function for even or odd

```python
## With out lambda function
def even_or_odd(num):
    if num%2==0:
        return 'even'
    else:
        return 'odd'

## calling
even_or_odd(33)
```

` Doing this with lambda function  `

```python
even_or_odd = lambda num: num%2==0 

# calling 
even_or_odd(22)
```

### **Or**

```python
even_or_odd = lambda num: 'even' if num%2==0 else 'odd'

# calling 
even_or_odd(22)
```

* Function that perform addition of Multiple Values 

```python
def add(x,y,z):
    return x+y+z

add(4,5,6)
```

`Addtion of multiple value with lambda function`

```python
add = lambda x,y,z:x+y+z

print(add(4,5,6))
```

### Lambda function has much more feature with map and filter function 

```python
### map()
numbers = [1,2,3,4,5,6]
# to find the square of each number
square_list=[]
def square_number(numbers):
    for i in numbers:
        square_list.append(i**2)
        
square_number(numbers)
print(square_list)
```

If we use `lambda function` for squaring a list of number with `map()` function

```python
list(map(lambda x:x**2,numbers))
```
