# Let's Start

```python
# for loop
#### For Loop is a loop it is used to iterate a sequence of numbers

for i in range(5):    # range(5) mean range(0,5) mean a sequence from 0-5 where 5 is not included
    print(i)
```

```python
for i in range(1,6):   # the range usually takes three parameters (start,stop,step) by default step=1
    print(i)
```

```python
for i in range(1,10,1):
    print(i)
```

```python
# step =2
for i in range(1,10,2):
    print(i)
```

```python
# i want all the numbers from 10 to 1
for i in range(10,1,-1):    # remember 1 is not included
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
# How to use for loop for string
str = "Shoaib M"

for i in str:
    print(i)
```

* If you have a collection of items or Numbers then you can easily traverse through it using **for loop**

## While Loop

* while loop continue to execute as long as the condition is true. Whatever the condition we are writing in while loop

```python
## While Loop 
# while loop continue to execute as long as the condition is true. What
# ever the condition we are writing in while loop
```

```python
count =0
while count<5:
    print(count)
    count= count + 1
```

```python
count =0
while count%2==0:
    print(count)
    count +=1
    
```

```python
## Loop Control Statement

## break
### The break exits the loop statement permaturely

for i in range(10):
    if i ==5:
        break       # to come out from the entire loop we use break statement
    print(i)
```

```python
## continue

## Continue statement skips the current iteration and continues with the next.
for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass
#### The pass statement is a null opearation; it does nothing

for i in range(6):
    if i==3:
        print('this value is going to pass')
        pass
    else:
        print(i)
```

# Nested Loops

```python
## Nested loops 
### Mean a loop inside another loop

for i in range(3):
    for j in range(2):
        print(i,j)
```

```python
# Q1 Calculate the sum of first N natural number using a while Loop
n= int(input("Enter a natual number: "))
count =1
sum=0
while count<=n:
    sum +=count
    count=count+1
print(f"The sum of first {n} number is: ",sum)
```

```python
# let's do the same thing with the help of for loop
n = 5
sum =0
for i in range(1,n+1):
    sum =sum+i
print(sum)
```

##### Prime Number 
* These are that number which are divisible by 1, or the number itself

```python

for num in range(1,101):
    if num > 1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
        
```
