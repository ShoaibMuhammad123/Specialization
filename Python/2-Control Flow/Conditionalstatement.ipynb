# **Conditional Statement**
A conditional statement in Python (or in programming generally) is a statement that executes certain actions only when a specific condition is true.

```python
## if statement --> execute when the condition is True
age = 20 
if age >=18:
    print("You are allowed to vote in election")
```

```python
## else statement 
if age<=18:
    print("You are not allowed to vote")
else:
    print("You can vote on your choise")
```

```python
## elif statement mean else if 
## it is used for multiple condition

marks =40
if marks <33:
    print("Fail")
elif marks <50:
    print("You just pass the exam")
else:
    print("congratulation you got")
```

```python
## Nested Conditional statement 
### when we write conditional statement inside another conditional statement that is known as nested if

# You can place one, or more if, elif, or else statements inside another if, elif, or else statements in a statement
gender = 'M'
age =30

if gender =="M":
    if age > 20:
        print("Mature Male Person")
    else:
        print("Not Mature")
else:
    if age >20: 
        print('Mature')
    else:
        print("Not Mature")
```

```python
### Practical Examples 

## Determine whether the year is a leap year 

year = int(input("Enter Year: "))

if (year % 4==0):
    if (year % 100 !=0):
        print(f"The year {year} is a Leap year")
    else:
        print("the year is not a Leap year")
else:
    print("Year is not a Leap year")
```

## **Simple Calculator**

```python
# take two input from user num1 and num2 
num1 = float(input("Enter First Number: "))
num2 = float(input("Enter Second Number: "))

# take an operator from user as input
operator = input("Enter operator (+,-,/,*): ")

# write conditions execute that only which match
if operator =="+":
    sum = num1 + num2
    print(sum)
elif operator =="-":
    diff = num1 - num2
    print(diff)
elif operator =="*":
    prod = num1/num2
    print(prod)
elif operator=="/":
    if num2!=0:
        quot = num1/num2
        print(quot)
    else:
        print('It gives an undefined number')
else:
    print("You enter incorrect Operator. Please try again")
print()
```
