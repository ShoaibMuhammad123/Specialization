#### **Why function would require?**

```python
# out program require to check whether a number is even or odd
# for this we write this code at once 

num = 24
if num%2==0:
    print("the number is even")
else:
    print("the number is odd")
```

##### The problem is when in other portion of my code if we have to check another number which is even or odd then again, I need to write the above code. this mean on requirement we have to write the code again and again multiple times **To solve this problem we use function**. In function we write a code only once and then use that code again and again by just calling it.

```python
## A better way of reusing the entire code is to use a Function, so we don't need to copy the code again and again

def even_or_odd(num):
    """This functions perform whether a number is even or odd"""
    if num%2==0:
        print("the number is even")
    else:
        print("the number is odd")
        
```

### `Positional Arguments`

```python
def print_numbers(*args):     # it start with asterick (*) args is conventional we can write *shoaib also 
    for numbers in args:
        print(numbers)
        

print_numbers(1,2,3,4,4,5)
```

* The `*shoaib` is refering to all the positional arguments

* `Positional Argument` mean that all the argument which are passing to the functions

```python
def print_numbers(*shoaib):     # it start with asterick (*) args is conventional we can write *shoaib also 
    for numbers in shoaib:
        print(numbers)
        

print_numbers(1,2,3,4,4,5,"my name is shoaib")
```

```python
## Positional Arguments

def addition(*args):  
    result=0   
    for numbers in args:
        result +=numbers
    return result

addition(1,2,3,4,4,5)
```

### `KeyWord Arguments`
* Keyword Argument start with double asterick  `**kwargs`
* In Keyword argument all the `parameter` will be in the form of `key-value pairs`
* It mean if it is in the form of key-value pair then it can be access in the way we access dictionary
* All these parameters `name="Shoaib",age="32",Gender = "Male"`  are referenced by **kwargs
* Kyword argument store all the pass parameter in the form of dictionary

```python
## Keyword arguments      ## It start with double asterick **kwargs

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Shoaib",age="32",Gender = "Male")
```

### Combine use of `Positional` and `Keyword Arguments`

```python
def print_details(*args,**kwargs):
    for numbers in args:
        print(f"Positional argument:",numbers)
        
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(3,4,5,6,name = "shoaib",age=23,education="Bachelor")
```

### `Return Statement`
* After the entire statement execute in a function a `return` statement return a result

```python
def multiply(a,b):
    return a * b
multiply(3,4)
```

```python
def mul(*args):
    print(args)
    print(type(args))
    
mul(3,4,5,6)
```

```python
def func(**kwargs):
    print(kwargs)
    print(type(kwargs))
    
func(name='shoaib',height=5.9,age=22)
```
