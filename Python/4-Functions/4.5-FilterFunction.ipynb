# **Filter() Function in Python**
* Filter function is used to filter items from a given list(or any iterable ) based on a specific condition

```python
## To filter out where items in list is even or odd
def even_or_odd(num):
    return True if num%2==0 else False

even_or_odd(5)
```

```python
## To filter out the given list item base on specific condition we use filter function

lst = [4,5,6,4,3,2,1]
list(filter(even_or_odd,lst))
```

`Filter with a Lambda function`

```python
## lambda function 
odd = lambda num: num%2!=0

odd(3)
```

```python
## filter with lambda function
list(filter(odd,lst))
```

```python
# condition is to get only number greater than 5
num = [1,2,3,4,5,6,7,8,9]
list(filter(lambda x: x>5,num))
```

`Filter with lambda function and Multiple condition`

```python
## the two condtion are number should be greater than 
even_and_greater_than_five = list(filter(lambda x: x>5 and x%2==0,num))

print(even_and_greater_than_five)
```

```python
# how do i apply filters on dictionaries

data =[
    {'name':'akif','age':33} ,
    {'name':'muiz','age':32} 
]

list(filter(lambda x: x['age']>32,data))
```

#### Conclusion
* The filter() function is a powerful tool for creating iterators that filter items out of an iterable based on a function. It is commonly used for data cleaning, filtering objects, and removing unwanted elements from lists. By mastering filter(), you can write more concise and efficient code for processing and manipulating collections in python.
